```python
#1st kaggle competition 
#author : Gyanendra Prakash


import pandas as pd

df = pd.read_csv("train.csv")


L=[[]]
for i in range(df.shape[0]):
  L.append([df.iloc[i]['date'],df.iloc[i]['country']+"_"+df.iloc[i]['store']+"_"+df.iloc[i]['product'],df.iloc[i]['num_sold']])

newdf= pd.DataFrame(L, columns =['date', 'item','sold'])




  

```

```python
newdf.drop(axis=1,index=0,inplace=True)



```
